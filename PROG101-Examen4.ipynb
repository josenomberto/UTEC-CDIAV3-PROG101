```python
import pandas as pd


### 1. Carga de datos

# Carga de dataset diabetes
print('\n1.1 Cargando dataset Diabetes')
url = 'https://raw.githubusercontent.com/plotly/datasets/master/diabetes.csv'
df_diabetes = pd.read_csv(url)
print('\n   ✅ Dataset diabetes.csv cargado con éxito')

# Muestra las 5 primeras filas del df_diabetes
print('\n1.2 Mostrando las 5 primeras filas')
print('\n')
print(df_diabetes.head(5))

# Muestra información de números de filas, columnas y tipo de columnas
print('\n1.3 Mostrando numero de filas, columnas y tipo de columnas')
print('\n')
#print(f'a). Número de filas: {len(df_diabetes)}')
print(f'a). Número de filas: {df_diabetes.shape[0]}')
print(f'b). Número de columnas: {df_diabetes.shape[1]}')
print(f'c). Tipos de datos de cada columna:\n{df_diabetes.dtypes}')
print('\n1.4 Cambiando dtype de columna Outcome')
df_diabetes['Outcome'] = df_diabetes['Outcome'].astype('category')
print('\n')
print(df_diabetes.info())

### 2. Filtrado de datos

# Crea df_filtrado con las filas con glucosa > 120
print('\n2.1 Creando df_filtrado con glucosa > 120')
print('\n   ✅ Dataframe creado con éxito')
df_filtrado = df_diabetes[(df_diabetes['Glucose'] > 120)].copy()

# Agrega nueva columna 'BMI_Category' con categorias dependiendo de los rangos de BMI
print('\n2.2 Agregando columna BMI_Category a df_filtrado')
df_filtrado['BMI_Category'] = df_filtrado['BMI'].apply(lambda bmi: 'Bajo' if bmi < 18.5 else
                                                                 'Normal' if bmi >= 18.5 and bmi < 25 else
                                                                 'Sobrepeso' if bmi >= 25 and bmi < 30 else
                                                                 'Obeso'
                                                                 )
print('\n   ✅ Columna agregada con éxito')
# Muestra las 5 primeras columnas de df_filtrado con nueva columna
print('\n2.3 Mostrando las 5 primeras filas de df_filtrado')
print('\n')
print(df_filtrado.head(5))


### 3. Carga y combinado de dataframes

# Carga de dataset age_groups
print('\n3.1 Cargando dataset Age Groups')
filename = 'age_groups.csv'
df_age_groups = pd.read_csv(filename)
print('\n   ✅ Dataset age_groups.csv cargado con éxito')

# Combinado de datasets
print('\n3.2 Combinando dataset df_diabetes con df_age_groups en columna Age')
df_combinado = pd.merge(df_diabetes, df_age_groups, on = 'Age', how = 'left')
print('\n   ✅ Datasets combinados con éxito')

# Muestra las 5 primeras filas del df_combinado
print('\n3.3 Mostrando las 5 primeras filas')
print('\n')
print(df_combinado.head(5))

### 4. Agrupado de dataframe

# Agrupado de dataframe por AgeGroup, promedio de Glucose y promedio BMI
print('\n4.1 Agrupado de dataframe por AgeGroup, promedio de Glucose y promedio BMI')
df_agrupado_ordenado = df_combinado.groupby(['AgeGroup']).agg({'Glucose': 'mean','Age': 'mean'}).sort_values(by='Glucose', ascending=False)
print('\n   ✅ Dataset agrupado con éxito')

# Muestra df_agrupado_ordenado
print('\n4.3 Mostrando Dataframe luego de agrupado y ordernado')
print('\n')
print(df_agrupado_ordenado)

### 5. Filtrado de dataframe df_combinado

# Filtrado y ordenado de df_combinado
print('\n5.1 Creando df_filtrado de df_combinado por AgreGroup y Outcome, y ordenado por BMI')
df_filtro_age_groups = df_combinado[(df_combinado['AgeGroup'] == 'Adulto') & (df_combinado['Outcome'] == 1)].sort_values(by='BMI', ascending=False)
print('\n   ✅ Dataframe filtrado y ordenado con éxito')

# Muestra las 5 primeras filas de nuevo dataframe
print('\n5.3 Mostrando las 5 primeras filas de nuevo dataframe')
print('\n')
print(df_filtro_age_groups.head(5)[['Age', 'Glucose', 'BMI', 'DiabetesPedigreeFunction']])

### 6. Tabla pivot de dataframe df_combinado

# Creado de pivot table
print('\n6.1 Creando pivot table')
df_pivot_table = pd.pivot_table(df_combinado, index = 'AgeGroup', columns = 'Outcome',
                                values = ['Glucose', 'BMI'], aggfunc = 'mean', observed=False)
print('\n   ✅ Pivot table creada con éxito')

# Muestra la pivot table
print('\n6.2 Mostrando la pivot table')
print('\n')
df_pivot_table

# Filtrado de filas donde los promedios de BMI es mayor a 30
print('\n6.3 Mostrando la pivot table para valores promedio de BMI mayor a 30')
df_combinado_bmi = df_combinado[(df_combinado['BMI'] > 30)]
df_pivot_table_filtered = pd.pivot_table(df_combinado_bmi, index = 'AgeGroup', columns = 'Outcome',
                                values = ['Glucose', 'BMI'], aggfunc = 'mean', observed=False)


print('\n')
df_pivot_table_filtered
```
